## **Removing Duplicate Rows in a Dataset**

```python
import pandas as pd
```

##### **Importing the data**

```python
df = pd.read_csv('Sample - Superstore Old.csv', encoding='cp1252')
```

##### **Checking for presence of duplicates (row-wise)**

```python
df.duplicated()
```

##### **Checking for presence of any duplicates**

```python
df.duplicated().any()
```

##### **Number of duplicates**

```python
df.duplicated().sum()
```

##### **Duplicate rows**

```python
df[df.duplicated()]
```

##### **Dropping duplicates**

```python
df_no_dup = df.drop_duplicates()
df_no_dup
```

##### **Checking for presence of any duplicates (after removal)**

```python
df_no_dup.duplicated().any()
```

##### **Number of duplicates (after removal)**

```python
df_no_dup.duplicated().sum()
```

##### **Assigning the data to 'Data (No Duplicates).csv' file**

```python
df_no_dup.to_csv('Data (No Duplicates).csv', index=False)
```
